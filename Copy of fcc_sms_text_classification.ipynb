```
# get data files
!wget https://cdn.freecodecamp.org/project-data/sms/train-data.tsv
!wget https://cdn.freecodecamp.org/project-data/sms/valid-data.tsv

train_file_path = "train-data.tsv"
test_file_path = "valid-data.tsv"
```

```
train_data=pd.read_csv(train_file_path,sep='\t')
test_data=pd.read_csv(test_file_path,sep='\t')
```

```
# Import label encoder
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
# Encode labels in column 'species'.
train_data['ham']= label_encoder.fit_transform(train_data['ham'])
test_data['ham']= label_encoder.fit_transform(test_data['ham'])

```

```
train_labels=train_data['ham'].to_numpy()
train_values=train_data['ahhhh...just woken up!had a bad dream about u tho,so i dont like u right now :) i didnt know anything about comedy night but i guess im up for it.'].to_numpy()
test_labels=test_data['ham'].to_numpy()
test_values=test_data['i am in hospital da. . i will return home in evening'].to_numpy()
```

```
from keras.preprocessing.text import Tokenizer
tokenizer= Tokenizer()
tokenizer.fit_on_texts(train_values)
sequences=tokenizer.texts_to_sequences(train_values)
sequences=keras.utils.pad_sequences(sequences,padding='post',maxlen=50)
test_sequences = tokenizer.texts_to_sequences(test_values)
test_sequences = keras.utils.pad_sequences(test_sequences, padding='post', maxlen=50)
```

```
tet_sequence= tokenizer.texts_to_sequences(["how are you doing today"])
tet_sequence=keras.utils.pad_sequences(tet_sequence, padding='post', maxlen=50)
```

```
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(4178, 128),
    tf.keras.layers.LSTM(64),
    tf.keras.layers.Dense(64, activation="sigmoid"),
     tf.keras.layers.Dense(1)
])
model.summary()
model.compile(loss="binary_crossentropy",optimizer="adam",metrics=['acc'])

history = model.fit(sequences, train_labels, epochs=10, validation_split=0.2)
```

```
model.evaluate(test_sequences,test_labels)
```

```
# function to predict messages based on model
# (should return list containing prediction and label, ex. [0.008318834938108921, 'ham'])
def predict_message(pred_text):
  input_sequence = tokenizer.texts_to_sequences([pred_text])
  input_sequence = keras.utils.pad_sequences(input_sequence, padding='post', maxlen=50)
  # Make the prediction
  predictions = model.predict(input_sequence)
  class_labels =['ham', 'spam']  # Replace with your actual class labels
  predicted_class = class_labels[int(predictions[0] > 0.5)]  # Adjust the threshold as needed
  prediction=[predictions[0],predicted_class]
  return (prediction)

pred_text = "how are you doing today?"

prediction = predict_message(pred_text)
print(prediction)
```
